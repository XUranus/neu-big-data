```python
#http://api.bilibili.com/archive_stat/stat?aid=9
    
import json
import requests
 
    
def require_video(video_id):
    headers = {"User_Agent" : "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.90 Safari/537.36"}
    url = "http://api.bilibili.com/archive_stat/stat?aid="+str(video_id)   #b站视频信息api
    print(url)
    VID_info = requests.get(url = url,headers=headers).json()
    if(VID_info["message"] == "0"):
        hot_video = VID_info["data"]["view"]
        if hot_video != "--":
            return {'succ':True,'data':VID_info}
        else:
            return {'succ':False}
    else:
        return {'succ':False}
 
for aid in range(1,100):
    obj = require_video(aid)
    print('running on av'+str(aid))
    if(obj['succ']):
        with open("./"+str(aid)+".json","w") as f:
                json.dump(obj,f)
```
