```python
import requests
import re
import os
import sys
import json
import csv

# B站API详情 https://github.com/Vespa314/bilibili-api/blob/master/api.md

# 评论用户及其信息
info_list = []
finished = 0

# 获取一个AV号视频下所有评论
def getAllCommentList(item):
    url = "http://api.bilibili.com/x/reply?type=1&oid=" + str(item) + "&pn=1&nohot=1&sort=0"
    r = requests.get(url)
    numtext = r.text
    json_text = json.loads(numtext)
    if json_text["code"]!=0:
        #print('get info:av'+str(item)+',not exist')
        return False
    commentsNum = json_text["data"]["page"]["count"]
    page = commentsNum // 20 + 1
    for n in range(1,page):
        url = "https://api.bilibili.com/x/v2/reply?jsonp=jsonp&pn="+str(n)+"&type=1&oid="+str(item)+"&sort=1&nohot=1"
        req = requests.get(url)
        text = req.text
        json_text_list = json.loads(text)
        for i in json_text_list["data"]["replies"]:
            info_list.append([i["member"]["mid"],i["content"]["message"]])
        #print('get info:av'+str(item)+',success')
        return True

# 保存评论文件为txt
def saveTxt(filename,filecontent):
    filename = str(filename) + ".csv"
    with open('data/'+filename, "a", encoding='utf-8') as f:
        for content in filecontent:
            csv_write = csv.writer(f)
            data_row = [content[0],content[1]]
            csv_write.writerow(data_row)
    #print("one file writed")

if __name__ == "__main__":
    # 视频AV号列表
    start = 100000
    end = 100100
    aid_list = range(start,end)
    for item in aid_list:
        info_list.clear()
        if getAllCommentList(item):
            saveTxt(item,info_list)
        finished += 1
        print(str(finished)+' completed,'+str(end-start-finished)+'remain')
    print('all finished')
```
